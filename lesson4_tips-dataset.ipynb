```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

#### The distribution of "total_bill" and "tip"

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

sns.distplot(tips['total_bill'], ax=ax1)
sns.distplot(tips['tip'], ax=ax2)

ax1.set_xlabel('total_bill')
ax2.set_xlabel('tip')

plt.show()
```

#### The histogram

```python
def draw_histogram(feature):
    plt.hist(tips[feature])
    
    plt.xlabel(feature)
    plt.ylabel('frequency')
    plt.title('Histogram')
    
    plt.show()
```

```python
draw_histogram('size')
```

```python
draw_histogram('sex')
```

#### Group by, pie chart

```python
def group_dataframe(group_feature, merge_feature):
    group_series = tips.groupby(group_feature)[merge_feature].mean()
    group_df = pd.DataFrame(group_series)
    
    return group_df
```

```python
def bar_chart(title, merge_feature, dataframe):
    fig, ax = plt.subplots()
    ax.bar(dataframe.index, dataframe[merge_feature])
    
    ax.set_title(title)
    plt.show()
```

```python
sex_df = group_dataframe('sex', 'tip')
sex_df 
```

```python
bar_chart('Tip/ Sex bar chart', 'tip', sex_df)
```

```python
smoker_bill_df = group_dataframe('smoker', 'total_bill')
smoker_bill_df
```

```python
bar_chart('Total bill/ Smoker bar chart', 'total_bill', smoker_bill_df)
```

```python
smoker_tip_df = group_dataframe('smoker', 'tip')
smoker_tip_df
```

```python
bar_chart('Tip/ Smoker bar chart', 'tip', smoker_tip_df)
```

```python
c = tips.groupby('sex')['smoker'].count()
sex_smoker_df = pd.DataFrame(c)
```

```python
sex_smoker_df
```

```python
fig, ax = plt.subplots()
ax.bar(sex_smoker_df.index, sex_smoker_df['smoker'])

ax.set_title('Smoker/ Sex bar chart')
plt.show()
```
