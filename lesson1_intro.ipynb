### Plot the graph of a function y = sin(2$\pi$x + $\pi$/6)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 100 linearly spaced numbers
x = np.linspace(-4,4,100)
# x
```

```python
y = np.sin(x*np.pi + np.pi/6)
# y
```

```python
fig, ax = plt.subplots()

# plot the function
ax.plot(x,y, color='r')
ax.grid()

# show the plot
plt.show()
```

### Plot the graph of a function y = 3x<sup>2</sup> + 1 with 100 points (The points are generated randomly based on normal distribution)

```python
# 100 linearly spaced numbers with normal distribution ()
x = np.random.normal(1,10,100)
print(np.mean(x))
print(np.std(x))
# print(x)
```

```python
y = 3*x**2 + 1
# print(y)
```

```python
fig, ax = plt.subplots()

# plot the function
ax.scatter(x,y, color='r')
ax.grid()

# show the plot
plt.show()
```
