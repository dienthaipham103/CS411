```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load data
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# plot distribution of 4 features
fig, ax = plt.subplots(4, 1, sharex=True)

# set the size of figure
fig.set_size_inches([6, 10])

ax[0].hist(iris['sepal_length'], color='b', bins=20)
ax[1].hist(iris['sepal_width'], color='r', bins=20)
ax[2].hist(iris['petal_length'], color='y', bins=20)
ax[3].hist(iris['petal_width'], color='c', bins=20)

ax[0].set_ylabel('sepal_length')
ax[1].set_ylabel('sepal_width')
ax[2].set_ylabel('petal_length')
ax[3].set_ylabel('petal_width')

ax[3].set_xlabel("cm")

plt.tight_layout()
plt.show()
```
